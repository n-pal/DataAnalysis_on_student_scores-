```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
import seaborn as sb
```

```python
df = pd.read_csv("student-scores.csv")
```

```python
df.isnull().sum()
```

```python
df = df.drop("id", axis=1)
```

```python
df.head()
```

```python
ax = sb.countplot(data = df, x = "gender")
ax.bar_label(ax.containers[0])
plt.figure(figsize = (2, 2))
plt.show()
```

```python
# from above chart numbers of females are more than males
```

```python
gb = df.groupby("career_aspiration").agg({"math_score" : 'mean', "physics_score" : 'mean', "chemistry_score" : 'mean', "biology_score" : 'mean'})
print(gb)
```

```python
sb.heatmap(gb, annot = True)
plt.show()
```

```python
# from the above chart we have conclude that the Doctors and Scientist have good impact on thier scores.
```

```python
ax = sb.countplot(data = df, x = "part_time_job")
ax.bar_label(ax.containers[0])
plt.figure(figsize = (2, 2))
plt.show()
```

```python
ax = sb.countplot(data = df, x = "extracurricular_activities")
ax.bar_label(ax.containers[0])
plt.figure(figsize = (2, 2))
plt.show()
```
